# Dataset

```python
from studioai.data.credit import CreditScoreDataset
from studioai.container import VisualContainer
from studioai.service.io import IOService
```

```python

container = VisualContainer()
container.init_resources()
container.wire(packages=["studioai"])
```

```python
fp = "data/Credit Score Classification Dataset.csv"
df = IOService.read(fp)
ds = CreditScoreDataset(df=df)
```

```python
ds.plot.lineplot(x='Income', y='Age', hue='Education')
ds.plot.lineplot(x='Income', y='Age')
```

```python
ds.plot.scatterplot(x='Income', y='Age', hue='Education')
ds.plot.scatterplot(x='Income', y='Age')
```

```python
ds.plot.histogram(x='Income', y='Age', hue='Education')
ds.plot.histogram(x='Income', y='Age')
ds.plot.histogram(x='Income', title='Histogram of Income')
```

```python
ds.plot.boxplot(x='Income', y='Education')
ds.plot.boxplot(x='Income', title='Histogram of Income')
```

```python
ds.plot.kdeplot(x='Income', y='Age')
ds.plot.kdeplot(x='Income', title='Probability Density of Income')
```

```python
ds.plot.ecdfplot(x='Income', title='Probability Density of Income')
```

```python
ds.plot.barplot(x='Income', y='Education')
ds.plot.barplot(x='Income', title='Probability Density of Income')
```

```python
ds.plot.violinplot(x='Income', y='Education')
ds.plot.violinplot(x='Income', title='Probability Density of Income')
```

```python
ds.plot.histpdfplot(x='Income', y='Education')
ds.plot.histpdfplot(x='Income', title='Probability Density of Income')
```

```python
ds.plot.pdfcdfplot(x='Income', title='Probability Density of Income')
```

```python
ds.plot.pairplot(vars=['Income', 'Education'])
ds.plot.pairplot(vars=['Income'], title='Probability Density of Income')
```

```python
ds.plot.jointplot(x='Income', y='Education', title="Pairplot")
```

```python
ds.plot.topn_plot(x='Income')
```
